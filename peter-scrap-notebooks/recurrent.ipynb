```python
df_intensity_1_0 = pd.read_csv("./matlab/outputdata/intensity_1-0.csv")
df_intensity_1_1 = pd.read_csv("./matlab/outputdata/intensity_1-1.csv")
df_intensity_1_2 = pd.read_csv("./matlab/outputdata/intensity_1-2.csv")
df_intensity_1_3 = pd.read_csv("./matlab/outputdata/intensity_1-3.csv")
df_intensity_1_4 = pd.read_csv("./matlab/outputdata/intensity_1-4.csv")
df_intensity_1_5 = pd.read_csv("./matlab/outputdata/intensity_1-5.csv")
df_intensity_1_6 = pd.read_csv("./matlab/outputdata/intensity_1-6.csv")
df_intensity_1_7 = pd.read_csv("./matlab/outputdata/intensity_1-7.csv")
df_intensity_1_8 = pd.read_csv("./matlab/outputdata/intensity_1-8.csv")
df_intensity_1_9 = pd.read_csv("./matlab/outputdata/intensity_1-9.csv")
df_intensity_2_0 = pd.read_csv("./matlab/outputdata/intensity_2-0.csv")
dframes = [df_intensity_1_0, df_intensity_1_1, df_intensity_1_2, df_intensity_1_3, df_intensity_1_4, df_intensity_1_5, df_intensity_1_6, df_intensity_1_7, df_intensity_1_8, df_intensity_1_9, df_intensity_2_0]
```

```python
for e, df in enumerate(dframes):
    plt.figure(figsize=(7, 2))
    plt.title('Input Signal {}uA Peak'.format(1+e*0.1))
    plt.plot(df['Input signal']+e*0.1)
    plt.xlabel('time')
    plt.ylabel('uA')
    plt.figure()
    t = df['time']
    plt.xlabel('time (ms)')
    plt.plot(t, df['Vd'], label='Vd')
    plt.plot(t, df['Vs'], label='Vs')
    plt.legend()
```
