```python
import scipy.io as sio
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (7,7)
```

```python
mat = sio.loadmat('./matlab/scrap/output.mat', squeeze_me=True)
```

```python
trials = mat['trial_output']
for e, t in enumerate(trials):
    peak = t[0]
    time = t[1]
    Vd = t[2]
    Vs = t[3]
    input_sig = t[4]

    fig, ax = plt.subplots()
    ax.plot(time, input_sig)
    ax.set_xlabel('time (ms)')
    ax.set_title('Input signal')

    fig, ax = plt.subplots()
    if e <= 2:
        ax.set_ylim([0, 4])
    ax.set_title('With input signal peak {}'.format(peak))
    ax.plot(time, Vd, label='Vd')
    ax.plot(time, Vs, label='Vs')
    ax.set_xlabel('time (ms)')
    ax.legend()
```
