# Methods

## Na


Dendritic voltage

When considering interactions with the soma, we will have 
$$
V_d(t) = \sigma(\kappa_1 \ast (I_d+S)) + \kappa_2 \ast (I_d + S)
$$
where S is the somatic spike (unit impulse filtered by a rectangular pulse of 1ms in width). 

For a passive propagation to soma we would have:
$$ V_s(t) = \kappa_s\ast I_d $$
with a threshold condition on $V_s$ for a somatic spike.
